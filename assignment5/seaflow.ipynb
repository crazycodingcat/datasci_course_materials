```python
from sklearn.cross_validation import train_test_split
```

```python
# df_train, df_test = train_test_split(df, test_size=0.5)
df_train, df_test = train_test_split(df[['fsc_small', 'fsc_perp', 'chl_small', 'pe', 'chl_big', 'chl_small', 'pop']], test_size=0.5)
```

```python
import seaborn as sns
%matplotlib inline
```

```python
sns.lmplot('pe','chl_small',data=df, hue='pop', fit_reg=False, size=8)
```

```python
from sklearn import tree
```

```python
clf = tree.DecisionTreeClassifier(max_depth=3)
```

```python
clf = clf.fit(df_train.iloc[:, :-1], df_train.iloc[:,-1])
```

```python
from sklearn.externals.six import StringIO
import pydot
```

```python
dot_data = StringIO()
tree.export_graphviz(clf, out_file=dot_data, max_depth=6,
                     feature_names=['fsc_small', 'fsc_perp', 'chl_small', 'pe', 'chl_big', 'chl_small'],  
                         class_names=df['pop'].unique(),  
                         filled=True, rounded=True,  
                         special_characters=True)
```

```python
graph = pydot.graph_from_dot_data(dot_data.getvalue())
graph.write_pdf("seaflow.pdf")
```
